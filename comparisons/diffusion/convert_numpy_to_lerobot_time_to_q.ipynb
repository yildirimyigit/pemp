```python
from pathlib import Path
import numpy as np

from lerobot.datasets.lerobot_dataset import LeRobotDataset
```

```python
data_npy = np.load("/home/yigit/projects/pemp/data/xarm/processed/norm_400.npy")
datasets_root = Path("/home/yigit/projects/pemp/comparisons/diffusion/lerobot_data/")

repo_id = "local/my_skill_time2q"
fps = 50  # set this to your actual control rate (Hz)
task_str = "my_skill"

# observation.state = timestep t (scalar), action = 7D joint positions
features = {
    "observation.state": {
        "dtype": "float32",
        "shape": (1,),
        "names": ["t"],
    },
    "action": {
        "dtype": "float32",
        "shape": (7,),
        "names": ["q1", "q2", "q3", "q4", "q5", "q6", "q7"],
    },
}

dataset_path = datasets_root / repo_id
dataset = LeRobotDataset.create(
    repo_id=repo_id,
    fps=fps,
    root=dataset_path,
    features=features,
    use_videos=False,
)
```

```python
for i, traj in enumerate(data_npy):
    T = traj.shape[0]

    for t in range(T):
        obs_t = np.array([t/T], dtype=np.float32)  # normalized time
        act_t = traj[t].astype(np.float32)  # joint positions

        dataset.add_frame({
            "observation.state": obs_t,
            "action": act_t,
            "task": task_str,
        })
    dataset.save_episode()
    print(f"Added trajectory {i}")
```
