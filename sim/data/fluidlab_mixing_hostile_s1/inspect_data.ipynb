```python
# Load all .npz under raw/
import numpy as np
import os

data_files = [f for f in os.listdir("raw/") if f.endswith(".npz")]
data = []
for file in data_files:
    data.append(np.load(os.path.join("raw/", file)))
num_demos = len(data)
print(f"Loaded {num_demos} demos.")
```

```python
print(data[0].keys())
```

```python
for k in range(num_demos):
    print(data[k]['n_loops'])
    print(data[k]['g'])
    print('---')
```

```python
max_val = -100
min_len = 1000
for k in range(num_demos):
    max_val = np.max(np.abs(data[k]['actions']))
    min_len = min(min_len, data[k]['actions'].shape[0])

print(max_val)
print(min_len)
```
