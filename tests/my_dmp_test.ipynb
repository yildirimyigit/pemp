```python
import matplotlib.pyplot as plt
import sys
import numpy as np


folder_path = '../models/'
if folder_path not in sys.path:
    sys.path.append(folder_path)

from dmp import DMP
dmp = DMP(n_bfs=30, spring=200.0, damping=12)

dmp.alpha = 0.15

t = np.linspace(0, 1, 100)
x_demo = np.sin(2 * np.pi * t)
dmp.fit(x_demo.reshape(-1, 1))
x_new, dx_new, ddx_new = dmp.generate(x_demo[0]+0.01, x_demo[-1], 0.01, 1.0)

# Plot the results
plt.plot(t, x_demo, label='Demo')
plt.plot(t, x_new[:len(t)], label='Generated')
plt.legend()
plt.show()

# Generate a new trajectory starting at a different point
# x0 = 0.5
# goal = 1.5 
# generated_x, _, _ = dmp.generate(x0, goal, dt, T)

# plt.plot(time, demo_x, label="Demonstration")
# plt.plot(time, generated_x, label="Generated")
# plt.xlabel("Time (s)")
# plt.ylabel("Position")
# plt.legend()

# plt.tight_layout()
# plt.show()
```
